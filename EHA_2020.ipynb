```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
import selenium.webdriver.support.expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
from tqdm.notebook import tqdm
import time
```

```python
presidentialSymposium = 'https://library.ehaweb.org/eha/#!*menu=6*browseby=7*sortby=2*media=3*ce_id=1766*marker=755'
oralPresentations = 'https://library.ehaweb.org/eha/#!*menu=6*browseby=7*sortby=2*media=3*ce_id=1766*marker=756'
ePosterPresentations = 'https://library.ehaweb.org/eha/#!*menu=6*browseby=7*sortby=2*media=3*ce_id=1766*marker=757'
publicationOnly = 'https://library.ehaweb.org/eha/#!*menu=6*browseby=7*sortby=2*media=3*ce_id=1766*marker=758'
```

```python
driver = webdriver.Chrome(r'C:\Users\venug\Downloads\chromedriver.exe')
```

```python
driver.get(presidentialSymposium)
```

```python
listing = driver.find_element(By.ID,'listing-main')
```

```python
abstractFolders = []
presentationLists = []
subLists = listing.find_elements(By.CLASS_NAME,'list-mb')
for i in subLists:
    abstractFolders.append(i.find_element(By.TAG_NAME,'a').get_attribute('href'))
```

```python
for j in abstractFolders:
    driver.get(j)
    timeout = 500
    locator = '//*[@id="listing-main"]/div'
    WebDriverWait(driver, timeout).until(EC.visibility_of_element_located((By.XPATH, locator)))
    presentations = driver.find_elements(By.CLASS_NAME,'list-conf')
    for i in presentations:
        presentationLists.append(i.get_attribute('href'))
```

```python
presentationLists
```

```python
dfPresidentialSymposium = pd.DataFrame()
for presentation in tqdm(presentationLists):
    driver.get(presentation)
    timeout = 500
    locator = '//*[@id="heading"]/div/div[1]'
    
    WebDriverWait(driver, timeout).until(EC.visibility_of_element_located((By.XPATH, locator)))
    data = driver.page_source
    data = BeautifulSoup(data,'lxml')
    rowDetails = {}
    titleAndAuthors = data.find('div',id = 'heading')
    try:
        rowDetails['Title'] = titleAndAuthors.select('div.details>div.title')[0].text
    except:
        pass

    try:
        rowDetails['Authors'] = titleAndAuthors.select('div.details>div.co-author-list')[0].text
    except:
        pass
    AbstractDetails = data.find_all('p')
    count = 0
    for detail in AbstractDetails:
        try:
            rowDetails[detail.find('b').text] = detail.text
        except:
            try:
                rowDetails['Unknown Detail ' + str(count)] = detail.text
                count+=1
            except:
                pass
    dfPresidentialSymposium = dfPresidentialSymposium.append(rowDetails,ignore_index = True)
```

```python
dfPresidentialSymposium['Authors'] = dfPresidentialSymposium['Authors'].str.strip().str.replace('\n','').str.replace('  ','')
```

```python
dfPresidentialSymposium.to_excel('EHAPresidentialSymposium.xls')
```
