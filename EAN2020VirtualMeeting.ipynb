```python
import requests
import pandas as pd
from tqdm.notebook import tqdm
```

```python
Dates = ['2020-05-23','2020-05-24','2020-05-25','2020-05-26']
```

```python
headers = {'Host': 'api-lb.virtual-meeting.net',
    'Connection': 'keep-alive',
    'Accept': 'application/json, text/javascript, */*; q=0.01',
    'Authorization': 'Basic m-events:kims',
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/81.0.4044.138 Safari/537.36','Origin': 'https://www.eanvirtualcongress.org',
    'Sec-Fetch-Site': 'cross-site',
    'Sec-Fetch-Mode': 'cors',
    'Sec-Fetch-Dest': 'empty',
    'Referer': 'https://www.eanvirtualcongress.org/',
    'Accept-Encoding': 'gzip, deflate, br'
    ,'Accept-Language': 'en-GB,en-US;q=0.9,en;q=0.8'}
```

```python
SessionIds = []
for j in range(len(Dates)):
    url = 'https://api-lb.virtual-meeting.net/v1/rooms.json'  # Request to get all events in a day
    
    params = {'per_page': '100',
    'event_id': '263',
    'include_sponsor': '1',
    'include_contentsessions': '1',
    'starts_at_from': Dates[j],
    'page': '1'}
    data = requests.get(url,headers = headers,params = params)
    SessionList = data.json()
    
    
    for i in SessionList:
        for j in i['contentsessions']:
            SessionIds.append(j['id'])
SessionIds
```

```python
# Request for a particular session
id_ = '46788'
url = 'https://api-lb.virtual-meeting.net/v1/contentsessions/'+id_+'.json?include_resourcegroups_body=1&include_tags=1'
sessionData = requests.get(url,headers = headers)
sessionData.json()
```

```python
dfDetails = pd.DataFrame()
for id_ in tqdm(SessionIds):
    url = 'https://api-lb.virtual-meeting.net/v1/contentsessions/'+str(id_)+'.json?include_resourcegroups_body=1&include_tags=1'
    sessionData = requests.get(url,headers = headers)
    sessData = sessionData.json()
    
    # Session Details
    SessionDetails = {}
    sessionTitle = sessData['title']
    startTime = sessData['starts_at']
    endTime = sessData['ends_at']
    description = sessData['description']
    chairs = sessData['chairs']
    sessionLocation = sessData['room']['title']
    sessionNumber = sessData['number']
    sessionType = sessData['contentsessiontype']['name']
    chairNames = []
    for i in chairs:
        chairNames.append(str(i['firstname']) + " " + str(i['surname']) + ', City: ' + str(i['city']) +', Institute: ' + str(i['institute']))
    presentations = sessData['resourcegroups']
    SessionDetails['Session Title'] = sessionTitle
    SessionDetails['Session Start Time'] = startTime
    SessionDetails['Session End Time'] = endTime
    SessionDetails['Session Description'] = description
    SessionDetails['Location'] = sessionLocation
    SessionDetails['Session Number'] = sessionNumber
    SessionDetails['Session Type'] = sessionType
    SessionDetails['Chairs'] = chairNames.copy()


    if(len(presentations) == 0):
        dfDetails = dfDetails.append(SessionDetails,ignore_index = True)
    else:
        for i in range(len(presentations)):
            presentation = presentations[i]['body']['resources'][0]
            presentationDetails = SessionDetails.copy()
            presentationDetails['Presentation Title'] = presentation['title']
            presentationDetails['Presentation Start Time'] = presentation['starts_at']
            presentationDetails['Presentation End Time'] = presentation['ends_at']
            presentationDetails['Presentation Abstract Id'] = presentation['abstract_id']

            presentationAuthors = presentation['authors']
            authorNames = []
            for i in presentationAuthors:
                authorNames.append(str(i['firstname']) + " " + str(i['surname']) + ', City: ' + str(i['city']) +', Institute: ' + str(i['institute']))
            presentationDetails['Presentation Authors'] = authorNames
            dfDetails = dfDetails.append(presentationDetails,ignore_index = True)
```

```python
dfDetails
```

```python
dfDetails['Chairs'] = dfDetails['Chairs'].astype(str).str.replace("[",'').str.replace("]",'').str.strip("'")
dfDetails['Presentation Authors'] = dfDetails['Presentation Authors'].astype(str).str.replace("[",'').str.replace("]",'').str.strip("'")
```

```python
dfDetails['Date'] = dfDetails['Session Start Time'].str.split("T",expand = True)[0]
```

```python
dfDetails['Presentation End Time'] = dfDetails['Presentation End Time'].str.split("T",expand = True)[1].str.replace('.000Z','')
```

```python
dfDetails['Presentation Start Time'] = dfDetails['Presentation Start Time'].str.split("T",expand = True)[1].str.replace('.000Z','')
```

```python
dfDetails['Session Start Time'] = dfDetails['Session Start Time'].str.split("T",expand = True)[1].str.replace('.000Z','')
dfDetails['Session End Time'] = dfDetails['Session End Time'].str.split("T",expand = True)[1].str.replace('.000Z','')
```
